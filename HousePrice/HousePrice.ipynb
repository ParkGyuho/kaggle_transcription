```python
#왜도는 양수이면 왼쪽에 쳐짐, 음수면 오른쪽에 쳐짐
print('왜도:',train['SalePrice'].skew())
#첨도는 양수면 뾰족, 음수면 완만..
print('첨도:',train['SalePrice'].kurt())
```
